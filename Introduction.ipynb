# Introduction to image processing
Images serve several purposes including conveying information, entertainment and even saving life in medical imaging applications. It has been said that "One picture is worth more than a thousand words".

## What is an image
We define an image as a two dimensional function $f(x,y)$ where $x$ and $y$ are spatial conditions and $f(x,y)$ denotes the intensity of the image at the point $(x,y)$. A digital image consists of a finite number of elements known as pixels. These pixels take on a finite number of intensity levels.


### Binary Images
In binary images, the pixels have two intensity levels. QR codes are an example of a binary image.

![](img/QR_Code_Example.svg)


### 8-bit grayscale images
Gray scale images have intensity values that take one of $2^8=256$ values. When a grayscale image is loaded into a software package, it is represented as a matrix whose entries take values from $0$ to $255$.

The image is represented as an $M\times N$ matrix whose elements are the values taken by $f(x,y)$. We have 

$
f(x,y) = \begin{bmatrix}
f(0,0) & f(0,1) & \ldots, f(0,N-1) \\
f(1,0) & f(1,1) & \ldots, f(1,N-1) \\
\vdots & \vdots& & \vdots \\
f(M-1,0) & f(M-1,1) & \ldots, f(M-1,N-1) \\
\end{bmatrix}
$
